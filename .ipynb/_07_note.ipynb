```
import numpy as np
import torch

#create a tensor
x=torch.Tensor(3,4)
y=torch.tensor([1,2,3])
print(x)
print(y)
print(x.type(),y.type())
print(x[1,2])
print(y[1])
```

```
#create a random tensor
x=torch.randn(2,3)
y=torch.rand(2,3)
print(x)
print(y)
```

```
#list
arr=torch.Tensor(np.random.rand(2,3))
x=torch.from_numpy(np.random.rand(2,3))
print(arr)
print(x)
```

```
print(arr.type())
print(x.type())
```

```
aa=torch.Tensor(3,)
bb=torch.Tensor(3,1)
aa_t=torch.t(aa)
bb_t=torch.t(bb)

cc=torch.Tensor([[1,2,3],[4,5,6]])
dd=torch.Tensor(2,3)

print("aa:{0},\naa_t:{1}".format(aa.size(),aa_t.size()))
print("bb:{0},\nbb_t:{1}".format(bb.size(),bb_t.size()))
print(cc.type(),dd.type())
print(cc)
print(dd)
```

```
aa=aa.reshape(-1,1)
aa_t=torch.t(aa)
print("aa:{0},\naa_t:{1}".format(aa.size(),aa_t.size()))
print(aa)
print(aa_t)
```

```
x=torch.Tensor(3,4)
y=x.reshape(2,6)
z=x.view(2,6)
print(x)
print(y,y.type())
print(z,z.type())
```

```
#gradient
x=torch.rand(3,4,requires_grad=True)
y=x**3+2
#y.backward()
```

```
print(x)
print("---------------------")
print(y)
```

```
z=y.mean()
print(z)
```

```
z.backward()
print(x.grad)
```

```
from torch.autograd import Variable
a=Variable(torch.FloatTensor([[2.,4.]]),requires_grad=True)
b=torch.zeros(1,2)
b=a**2+1
out=2*b
out.backward(torch.FloatTensor([[1.,1.]]))
print("---------------------")
print(a.data)
print("---------------------")
print(out.data)
print("---------------------")
print(a.grad)
```

```
print(a.shape)
```

```
#gradient
aa=Variable(torch.FloatTensor([[33,44],[55,66],[6,7]]),requires_grad=True)
bb=a*3+2
bb.backward(torch.FloatTensor([[3,4],[5,5],[6,7]])) #传入的shape维度不能超过aa的shape
print(aa.data)
print(bb)
print(aa.grad)
```
